### Desenvolvimento Tabelas de **Processamento**

```python
import matplotlib.pyplot as plt
import pandas as pd


# Carregar os dados dos arquivos CSV
df_processa_americana = pd.read_csv('ProcessaAmericanas.csv', sep=';')
df_processa_mesa = pd.read_csv('ProcessaMesa.csv', sep=';')
df_processa_sem = pd.read_csv('ProcessaSemclass.csv', sep=';')
df_processa_viniferas = pd.read_csv('ProcessaViniferas.csv', sep=';')

# Agrupar todas as tabelas e um unico DF
df_processa_tinta_brancos = pd.concat([df_processa_americana, df_processa_mesa, df_processa_viniferas])
df_processa_viniferas2 = pd.concat([df_processa_viniferas])
df_processa_mesa2 = pd.concat([df_processa_mesa])
df_processa_americana2 = pd.concat([df_processa_americana])

# Analise 1 => Apenas Tintas e Brancas Rosadas tabelas agrupadas
df_processa_n1_tinta_brancos = df_processa_tinta_brancos[(df_processa_tinta_brancos.cultivar !='TINTAS')]
df_processa_n1_tinta_brancos = df_processa_n1_tinta_brancos[(df_processa_n1_tinta_brancos.cultivar !='BRANCAS E ROSADAS')]


# Analise 2 => Exceto Tintas e Brancas Rosadas tabelas agrupadas
df_processa_tinta_brancos = df_processa_tinta_brancos[df_processa_tinta_brancos['cultivar'].str.contains('^(TINTAS|BRANCAS E ROSADAS)$', regex=True)]


# Analise 3 => Apenas  Tintas e Brancas Rosadas tabela Viniferas
df_processa_viniferas = df_processa_viniferas[df_processa_viniferas['cultivar'].str.contains('^(TINTAS|BRANCAS E ROSADAS)$', regex=True)]
df_processa_viniferas = df_processa_viniferas[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]


#Analise 4 => Exceto Tintas e Brancas Rosadas tabela Viniferas
df_processa_viniferas2 = df_processa_viniferas2[(df_processa_viniferas2.cultivar !='TINTAS')]
df_processa_viniferas2 = df_processa_viniferas2[(df_processa_viniferas2.cultivar !='BRANCAS E ROSADAS')]

#Analise 5 => Apenas  Tintas e Brancas Rosadas tabela Mesa
df_processa_mesa = df_processa_mesa[df_processa_mesa['cultivar'].str.contains('^(TINTAS|BRANCAS)$', regex=True)]
df_processa_mesa = df_processa_mesa[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]


#Analise 6 => Exceto Tintas e Brancas Rosadas tabela Mesa
df_processa_mesa2 = df_processa_mesa2[(df_processa_mesa2.cultivar !='TINTAS')]
df_processa_mesa2 = df_processa_mesa2[(df_processa_mesa2.cultivar !='BRANCAS')]


#Analise 7 => Apenas  Tintas e Brancas Rosadas tabela Americanas
df_processa_americana = df_processa_americana[df_processa_americana['cultivar'].str.contains('^(TINTAS|BRANCAS E ROSADAS)$', regex=True)]
df_processa_americana = df_processa_americana[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]


#Analise 8 => Exceto Tintas e Brancas Rosadas tabela Americanas
df_processa_americana2 = df_processa_americana2[(df_processa_americana2.cultivar !='TINTAS')]
df_processa_americana2 = df_processa_americana2[(df_processa_americana2.cultivar !='BRANCAS E ROSADAS')]


# Filtro dos Anos - Apenas os ultimos 15
df_processa_tinta_brancos = df_processa_tinta_brancos[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
df_processa_n1_tinta_brancos = df_processa_n1_tinta_brancos[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
df_processa_viniferas2 = df_processa_viniferas2[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
df_processa_mesa2 = df_processa_mesa2[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
df_processa_americana2 = df_processa_americana2[['cultivar','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]


# Agrupar as uvas por cultivar
df_processa_agrupado = df_processa_tinta_brancos.groupby('cultivar').sum()
df_processa_n_agrupados = df_processa_n1_tinta_brancos.groupby('cultivar').sum()
dfViniferasAgrupados = df_processa_viniferas.groupby('cultivar').sum()
dfMesaAgrupados = df_processa_mesa.groupby('cultivar').sum()
dfAmericanasAgrupados = df_processa_americana.groupby('cultivar').sum()
dfViniferas_n_Agrupados = df_processa_viniferas2.groupby('cultivar').sum()
dfMesa_n_Agrupados = df_processa_mesa2.groupby('cultivar').sum()
dfAmericanas_n_Agrupados = df_processa_americana2.groupby('cultivar').sum()

# Conseguir somar e trazer apenas as maiores 3 uvas
df_processa_n_agrupados = df_processa_n_agrupados.transpose()
vlr = df_processa_n_agrupados.sum()
df_processa_n_agrupados = df_processa_n_agrupados[vlr.sort_values(ascending=False).index[:3]]
df_processa_n_agrupados = df_processa_n_agrupados.transpose()

dfViniferas_n_Agrupados = dfViniferas_n_Agrupados.transpose()
vlr2 = dfViniferas_n_Agrupados.sum()
dfViniferas_n_Agrupados = dfViniferas_n_Agrupados[vlr2.sort_values(ascending=False).index[:5]]
dfViniferas_n_Agrupados = dfViniferas_n_Agrupados.transpose()

dfMesa_n_Agrupados = dfMesa_n_Agrupados.transpose()
vlr3 = dfMesa_n_Agrupados.sum()
dfMesa_n_Agrupados = dfMesa_n_Agrupados[vlr3.sort_values(ascending=False).index[:5]]
dfMesa_n_Agrupados = dfMesa_n_Agrupados.transpose()

dfAmericanas_n_Agrupados = dfAmericanas_n_Agrupados.transpose()
vlr4 = dfAmericanas_n_Agrupados.sum()
dfAmericanas_n_Agrupados = dfAmericanas_n_Agrupados[vlr4.sort_values(ascending=False).index[:5]]
dfAmericanas_n_Agrupados = dfAmericanas_n_Agrupados.transpose()

# Transpor o DataFrame
df_processa_agrupado = df_processa_agrupado.transpose()
df_processa_n_agrupados = df_processa_n_agrupados.transpose()
dfViniferasAgrupados = dfViniferasAgrupados.transpose()
dfMesaAgrupados = dfMesaAgrupados.transpose()
dfAmericanasAgrupados = dfAmericanasAgrupados.transpose()
dfViniferas_n_Agrupados = dfViniferas_n_Agrupados.transpose()
dfMesa_n_Agrupados = dfMesa_n_Agrupados.transpose()
dfAmericanas_n_Agrupados = dfAmericanas_n_Agrupados.transpose()

```
